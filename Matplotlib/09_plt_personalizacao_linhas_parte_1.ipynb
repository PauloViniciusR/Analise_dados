```python
import matplotlib.pyplot as plt
from cycler import cycler
import numpy as np

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
marcadores = np.where(vendas == np.min(vendas), "x", "")

# Paleta de cores
cores = plt.get_cmap("Accent").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)

# Gráfico base
plt.plot(meses, vendas)

# Marcadores e anotação
for posicao, marcador in enumerate(marcadores):
    if marcador == "x":
        cor="red"
    else:
        cor="blue"
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, color=cor)


plt.annotate("Menor valor", color="red", xy=(2, 78), xytext=(3, 80))

# Destaque de região
plt.axvspan(6 , 11, alpha=0.2, color=cores[1])

plt.text(6, 80, "Recuperação de crescimento", color="green", fontsize=9, weight="bold")

plt.xlabel("Mês")
plt.ylabel("Vendas (Em bilhões de doláres)")
plt.title("Crescimento de vendas ao longo do ano")


plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler
import numpy as np

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
marcadores = np.where(vendas == np.min(vendas), "x", "")

# Paleta de cores
cores = plt.get_cmap("Accent").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)

# Gráfico base
plt.plot(meses, vendas)

# Marcadores e anotação
for posicao, marcador in enumerate(marcadores):
    if marcador == "x":
        cor="red"
    else:
        cor=cores[7]
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, color=cor)
    plt.annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-5, 5), textcoords="offset points")

#plt.annotate("Menor valor", color="red", xy=(2, 78), xytext=(3, 80))

# Destaque de região
plt.axvspan(6 , 11, alpha=0.2, color=cores[1])

plt.text(6, 80, "Recuperação de crescimento", color="green", fontsize=9, weight="bold")

plt.box(False)
plt.tick_params(axis="x", length=0)
plt.yticks([])

plt.ylabel("Vendas (Em bilhões de doláres)")
plt.title("Crescimento de vendas ao longo do ano")


plt.show()
```
