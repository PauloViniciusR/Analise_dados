```python
import matplotlib.pyplot as plt
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.plot(meses, vendas)

cores = plt.get_cmap("Accent").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)

plt.annotate("Menor valor", color="red", xy=(2, 78), xytext=(3, 80), arrowprops=dict(facecolor="red"))

plt.xlabel("Mês")
plt.ylabel("Vendas (Em bilhões de doláres)")
plt.title("Crescimento de vendas ao longo do ano")


plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler
import numpy as np

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

marcadores = np.where(vendas == np.min(vendas), "x", "")

plt.plot(meses, vendas)

for posicao, marcador in enumerate(marcadores):
    if marcador == "x":
        cor="red"
    else:
        cor="blue"
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, color=cor)


cores = plt.get_cmap("Accent").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)

#plt.annotate("Menor valor", color="red", xy=(2, 78), xytext=(3, 80), arrowprops=dict(facecolor="red"))

plt.xlabel("Mês")
plt.ylabel("Vendas (Em bilhões de doláres)")
plt.title("Crescimento de vendas ao longo do ano")


plt.show()
```
