## Gráfico de barras para apresentação

Considere que você tem as quantidades de produtos vendidos em uma loja. Faça um gráfico de barras horizontais.

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

plt.bar(produtos, quantidades)

plt.show()
```

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=quantidades, padding=3, fontsize=10, fontweight="bold")

plt.box(False)

plt.xticks([])
plt.tick_params(axis="y", length=0)

plt.title("Quantidade de produtos vendidos")

plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler


produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
rotulos = [valor if valor == max(quantidades) else "" for valor in quantidades]

cores = plt.get_cmap("Pastel2").colors
cicle_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=cicle_cores)

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=rotulos, padding=3, fontsize=10, fontweight="bold")

plt.box(False)

plt.xticks([])
plt.tick_params(axis="y", length=0)

for i, barra in enumerate(barras):
    if i != quantidades.index(max(quantidades)):    
        barras[i].set_color(cores[1])
        barras[i].set_alpha(0.3)


plt.title("Quantidade de produtos vendidos")

plt.show()
```
