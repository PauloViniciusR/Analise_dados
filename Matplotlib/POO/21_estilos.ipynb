## Estilos Pré-Prontos

Estilos no Matplotlib são essencialmente temas que você pode aplicar aos seus gráficos para alterar sua aparência visual. Eles podem ser extremamente úteis quando você quer criar gráficos visualmente agradáveis de maneira rápida e eficiente.

Cada estilo é uma coleção de configurações de aparência que definem elementos como cores de fundo, cores de linha, estilos de linha, cores de marcador, estilos de fonte, entre outros. Ao aplicar um estilo, todas essas configurações são alteradas de uma vez, permitindo que você mude a aparência completa do seu gráfico com uma única linha de código.

O Matplotlib vem com uma variedade de estilos pré-prontos que você pode usar. Esses estilos são projetados para serem visualmente agradáveis e para funcionar bem em uma ampla gama de casos de uso.



```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=quantidades, padding=3, fontsize=10, fontweight='bold')

plt.xlabel('Quantidade Vendida')
plt.ylabel('Produto')
plt.title('Quantidade de Produtos Vendidos')

plt.show()
```

## Como Usar Estilos Pré-Prontos

Usar um estilo pré-pronto no Matplotlib é muito simples. Tudo o que você precisa fazer é chamar a função `plt.style.use()` e passar o nome do estilo que você deseja usar como argumento.

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=quantidades, padding=3, fontsize=10, fontweight='bold')

plt.xlabel('Quantidade Vendida')
plt.ylabel('Produto')
plt.title('Quantidade de Produtos Vendidos')

plt.show()
```
