```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

#----------------------------- gráfico 1 - linha -------------------------------------
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

axs[0,0].plot(meses, vendas)
axs[0,0].set_xlabel("Mês")
axs[0,0].set_ylabel("Vendas (Em milhares de reais)")
axs[0,0].set_title("Evolução de vendas ao longo do ano")

# ----------------------------- gráfico 2 - barra -------------------------------------
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

axs[0,1].bar(produtos, quantidades)
axs[0,1].set_xlabel("Mês")
axs[0,1].set_ylabel("Vendas (Em milhares de reais)")
axs[0,1].set_title("Evolução de vendas ao longo do ano")

# ----------------------------- gráfico 3 - pizza -------------------------------------
categorias = ["Eletrônicos", "Roupas", "Alimentos", "Livros"]
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # Destacar primeira fatia

axs[1,0].pie(quantidades, labels=categorias, explode=destaque, autopct="%.1f")
axs[1,0].set_title("Categorias produtos vendidos")

#------------------------------------- gráfico 4 - barras empilhadas -------------------------------------
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]


axs[1,1].bar(anos, vendas_online, label="Online")
axs[1,1].bar(anos, vendas_loja, bottom=vendas_online, label="Loja")
axs[1,1].set_xlabel("Ano") 
axs[1,1].set_ylabel("Vendas (Em milhões de reais)")
axs[1,1].set_title("Vendas online vs Loja")
axs[1,1].set_xticks(anos)
axs[1,1].legend()
#------------------------------------- ajustes finais -------------------------------------
fig.suptitle("Evolução de vendas", fontsize=16)


plt.show()
```
