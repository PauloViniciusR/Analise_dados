```python
import pandas as pd
```

```python
BASE = "dados/employee_attrition.csv"

df = pd.read_csv(BASE)
df.head()
```

```python
df.groupby("Gender").size()
```

```python
df.groupby("Gender")["Age"].value_counts(bins=10)
```

```python
df.groupby("Gender")["Age"].value_counts(bins=10).sort_index()
```

```python
df.groupby("Gender")["Age"].plot(kind="hist", legend=True, alpha=0.5)
```

```python
df.hist(column="Age", by="Gender")
```

```python
df.hist(column="Age", by="Gender", sharey=True)
```

```python
df.hist(column="Age", by="Gender", sharex=True, layout=(2,1), figsize=(5, 7))
```

```python
df.groupby("Attrition")["Age"].mean()
```

```python
df.groupby("Attrition")["Age"].value_counts(bins=10, normalize=True).mul(100).round(1).sort_index()
```

```python
df.hist(column="Age", by="Attrition", bins=10)
```
