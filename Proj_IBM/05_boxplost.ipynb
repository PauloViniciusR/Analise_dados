```python
import pandas as pd
```

```python
BASE = "dados/employee_attrition.csv"

df = pd.read_csv(BASE)
df.head()

```

```python
df["DistanceFromHome"].hist()
```

```python
df.boxplot(column="DistanceFromHome", vert=False, grid=False, showmeans=True)
```

```python
df["DistanceFromHome"].describe()
```

```python
df.groupby(["Gender","Attrition"])["Age"].describe()
```

```python
df.boxplot(column="Age", by=["Gender", "Attrition"], grid=False, showmeans=True)  
```

```python
df.groupby("JobLevel")["Age"].describe()
```

```python
df.groupby(["Gender", "Attrition"])["JobSatisfaction"].describe()
```

```python
df.boxplot(column="JobSatisfaction", by=["Gender", "Attrition"], grid=False, showmeans=True)
```

```python
df.boxplot(column="Age", by="JobLevel", grid=False, showmeans=True)
```

```python
df.groupby(["JobRole", "Gender", "JobLevel"])["MonthlyIncome"].describe()
```

```python
job_roles = df["JobRole"].unique()

job_roles
```

```python
for job in job_roles:
    df[df["JobRole"] == job].boxplot(column="MonthlyIncome", by=["JobLevel", "Gender"], grid=False, showmeans=True)
```

```python
for job in job_roles:
    df[df["JobRole"] == job].plot(kind="box", column="MonthlyIncome", by=["JobLevel", "Gender"], grid=False, showmeans=True, title=job)
```
