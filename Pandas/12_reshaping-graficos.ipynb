```python
import pandas as pd
```

```python
BASE = "dados/cereal_sem_negativos.csv"

df = pd.read_csv(BASE)
df.head()
```

```python
df.info()
```

```python
df.groupby(["mfr", "shelf"]).size()
```

```python
df.groupby(["mfr", "shelf"]).size().plot(kind="bar")
```

```python
df.groupby(["mfr", "shelf"]).size().unstack().plot(kind="bar")
```

```python
df.groupby(["mfr", "shelf"]).size().unstack().plot(kind="bar", stacked=True)
```

```python
# As funções stack() e unstack() são operações muito úteis no Pandas para manipulação de dados em um DataFrame, 
# especialmente quando queremos trabalhar com dados que têm múltiplos níveis de indexação.

# O stack() transforma um DataFrame em uma estrutura mais compacta, 
# fazendo o empilhamento dos níveis de colunas em um único nível de índice. Em outras palavras, ele "empilha" as colunas e as transforma em linhas.

# O unstack() faz o oposto do stack(). Ele "desempilha" os dados de um índice (nível de linha) e transforma em colunas. Usualmente, você aplica unstack() em um DataFrame que possui um índice de múltiplos níveis.
#Ele é útil quando queremos transformar os dados de modo que níveis de índices se tornem colunas.
```
