```python
import pandas as pd
```

```python
df = pd.read_excel("dados/reshaping.xlsx")
df
```

```python
df.pivot_table(
    index=["Estudante"],
    columns=["Semana", "Tópico"],
    values="Nota",
    aggfunc="sum"
)

```

```python
df.groupby(["Estudante", "Semana", "Tópico"])["Nota"].mean()
```

```python
df.groupby(["Estudante", "Semana", "Tópico"])["Nota"].mean().unstack(level=[1, 2])
```
