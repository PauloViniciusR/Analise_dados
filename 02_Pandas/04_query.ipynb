```python
import pandas as pd

BASE = "dados/cereal.csv"
```

```python
df = pd.read_csv(BASE)
df
```

```python
df[df["rating"] > 70][["name", "rating"]] #Opção1

# Rescreevendo com a query

df.query("rating > 70")[["name", "rating"]] #Opção2
```

```python
df[df["name"] == "Almond Delight"] #Opção1

# Rescreevendo com a query

df.query("name == 'Almond Delight'") #Opção2
```

```python
df[df["mfr"] == "N"] #Opção1

# Rescreevendo com a query

df.query("mfr == 'N'") #Opção2
```

```python
df[(df["mfr"] == "N") & (df["type"] == "H")] #Opção1

# Rescreevendo com a query

df.query("mfr == 'N' and type == 'H'") #Opção2
```

```python
df[(df["mfr"] == "N") & ~(df["type"] == "H")] #Opção1

# Rescreevendo com a query

df.query("mfr == 'N' and not type == 'H'") #Opção2
```

```python
df[(df["mfr"] == "N") | (df["type"] == "H")] #Opção1

# Rescreevendo com a query

df.query("mfr == 'N' or type == 'H'") #Opção2

```

```python
df[df['mfr'].isin(["N", "G", "N"])] #Opção1

# Rescreevendo com a query

df.query("mfr in ['N', 'G', 'A']") #Opção2
```

```python
df[(df["carbo"] == -1) | (df["sugars"] == -1) | (df["potass"] == -1)] #Opção1

# Rescreevendo com a query

df.query("carbo == -1 or sugars == -1 or potass == -1")
```
