```python
import pandas as pd

BASE = "dados/cereal_sem_negativos.csv"

df = pd.read_csv(BASE)
df.head()
```

```python
df["mfr"].value_counts().plot(kind="bar")
```

```python
df["mfr"].value_counts().plot(
    kind="bar",
    xlabel="Fabricante",
    ylabel="Quantidade",
    title="Contagem de cereais por fabricante"
)
```

```python
df["mfr"].value_counts().plot(
    kind="pie",
    xlabel="Quantidade",
    autopct= "%.1f%%",
    title="Percentual por fabricante",
) 
```

```python
df.groupby("mfr")["calories"].mean().plot(
    kind="bar",
    xlabel="Fabricante",
    ylabel="Calorias",
    title="Média de caloria por fabricante"
)
```

```python
df.groupby("mfr")["calories"].mean().plot(
    kind="line",
    xlabel="Fabricante",
    ylabel="Calorias",
    title="Média de caloria por fabricante"
)
```
